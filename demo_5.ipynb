```python
#Import libs
from src.core.annotate import ImageAnnotater as IA
from src.core.utils import utils

from PIL import Image

import os
import matplotlib.pyplot as plt
import numpy
```

```python
#Instantiate
ia = IA()
```

```python
#Open Image and convert to array
img1_ = Image.open('misc/nyc_single.jpg')
img1 = numpy.array(img1_)
```

```python
#Run inference
output = ia.predict(img1)
```

```python
#Extract words from outupt and join together to form a sentence
annotations = output[0]

#Form sentence
" ".join(annotations[1:-1]) #omit <start> and <end> tags
```

```python
#Check out what we've done!
ia.visualize(img1_,output)
```
